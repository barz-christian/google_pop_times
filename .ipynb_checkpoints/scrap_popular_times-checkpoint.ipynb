generalize

```python
def pop_times(x):
    # get url
    url = places_dic[x]
    # write url to file
    f = open('url.csv','w')
    f.write(url) 
    f.close()
    # scrap html site
    %run scrape_gm.py "url.csv"
    # load scraped info
    scrapped = pd.read_csv("sample_output.csv")
    # add place to data frame
    scrapped['place'] = x
    
    # rename
    scrapped = scrapped.rename(columns={"day_of_week": "day", 
                                        "hour_of_day": "hour", 
                                        "popularity_percent_normal": "pop",
                                        "popularity_percent_current": "pop_now"
                                       })
    
    # fill NaN
    scrapped = scrapped.fillna(0)
    
    return scrapped
```

```python
pop_times_places = [pop_times(x) for x in places]
```

## visualize

```python
def pop_places_visualise(x):
    df_extracted = x
    
    # extract place
    place = df_extracted['place'].unique()[0]
    
    # pivot table
    df_pivot = df_extracted.pivot_table(index="hour", columns="day", values ="pop")
       
    # plot pop times
    df_pivot.plot(title = place)
    
    return df_pivot


def pop_places_pivottable(x):
    df_extracted = x
    
    # extract place
    place = df_extracted['place'].unique()[0]
    
    # pivot table
    df_pivot = df_extracted.pivot_table(index="hour", columns="day", values ="pop")
    df_pivot['place'] = place
    
    return df_pivot
```

```python
popular_plots = [pop_places_visualise(x) for x in pop_times_places]
```

## write to csv for shiny

```python
# write pivot tables
pop_times_df = [pop_places_pivottable(x) for x in pop_times_places]
df_pop_times_pivot = pd.concat(pop_times_df)
df_pop_times_pivot.to_csv("pop_times_pivot.csv", sep=',')
df_pop_times_pivot.to_csv("./shiny/popular_places/data/pop_times_pivot.csv", sep=',')
```

```python
# write long tables (raw)
df_pop_times = pd.concat(pop_times_places)
df_pop_times.to_csv("pop_times.csv", sep=',')
df_pop_times.to_csv("./shiny/popular_places/data/pop_times.csv", sep=',')
```
